```python
import polars as pl
```

```python
df = pl.read_csv('scraping/assets/dogs_cleaned_with_descriptions.csv')
```

```python
for link in df["Detailed Description Link"]:
    print(link)
```

```python
import requests
import time
from scrapy import Selector

CONTENT_DIV = "div.entry-content.wp-block-post-content"
HEADERS = {"User-Agent": "Mozilla/5.0 (compatible; dogbreed-scraper/1.0)"}

descriptions = []
links = df["Detailed Description Link"]
total = len(links)

for i, link in enumerate(links, start=1):
    try:
        response = requests.get(link, headers=HEADERS, timeout=10)
        response.raise_for_status()
        sel = Selector(text=response.text)
        paragraphs = [
            " ".join(p.css("::text").getall()).strip()
            for p in sel.css(f"{CONTENT_DIV} > p")
        ]
        description = " ".join(p for p in paragraphs if p)
    except Exception as e:
        print(f"[{i}/{total}] Failed {link}: {e}")
        description = None

    descriptions.append(description)
    print(f"{i}/{total}")
    time.sleep(0.05)

df = df.with_columns(pl.Series("Description", descriptions))
```

```python
image_urls = []
links = df["Detailed Description Link"]
total = len(links)

for i, link in enumerate(links, start=1):
    try:
        response = requests.get(link, headers=HEADERS, timeout=10)
        response.raise_for_status()
        sel = Selector(text=response.text)
        image_url = sel.css("img.attachment-post-thumbnail.size-post-thumbnail.wp-post-image::attr(src)").get()
    except Exception as e:
        print(f"[{i}/{total}] Failed {link}: {e}")
        image_url = None

    image_urls.append(image_url)
    print(f"{i}/{total}")
    time.sleep(0.05)

df = df.with_columns(pl.Series("Image URL", image_urls))
```
